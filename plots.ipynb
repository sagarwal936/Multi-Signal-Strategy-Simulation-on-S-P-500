```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

## 0. Benchmark Strategy

```python
# Load and parse dates properly
BenchmarkStrategy = pd.read_csv(
    "output/BenchmarkStrategy.csv",
    parse_dates=["Date"]  # <-- ensures Date is datetime
)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(BenchmarkStrategy["Date"], BenchmarkStrategy["Cash"], label="Cash Balance")
plt.plot(BenchmarkStrategy["Date"], BenchmarkStrategy["Holdings"], label="Holdings")
plt.plot(BenchmarkStrategy["Date"], BenchmarkStrategy["Total Assets"], label="Total Assets")

# Format x-axis for time
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())             # major tick = every year
ax.xaxis.set_minor_locator(mdates.MonthLocator(interval=3))  # minor tick = every 3 months
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))  # e.g. 2022-03

plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Cash")
plt.title("Benchmark Strategy")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
